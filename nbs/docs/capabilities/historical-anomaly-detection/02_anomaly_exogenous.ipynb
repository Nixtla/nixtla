To detect anomalies with exogenous variables, load a dataset with the exogenous features as columns and use the same `detect_anomalies` method.

```python
#| hide
if not IN_COLAB:
    from dotenv import load_dotenv
```

```python
#| hide
load_dotenv()
nixtla_client = NixtlaClient()
```

```python
# Read the dataset
# The dataset has exogenous features in its columns
df = pd.read_csv('https://raw.githubusercontent.com/Nixtla/transfer-learning-time-series/main/datasets/electricity-short-with-ex-vars.csv')

# Detect anomalies
anomalies_df = nixtla_client.detect_anomalies(
    df=df,
    time_col='ds',
    target_col='y',
)

# Plot anomalies
nixtla_client.plot(df, anomalies_df)
```

```python
# Plot weights of date features
nixtla_client.weights_x.plot.barh(x='features', y='weights')
```
