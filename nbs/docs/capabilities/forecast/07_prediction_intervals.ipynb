## Step 2: Load dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Nixtla/transfer-learning-time-series/main/datasets/air_passengers.csv")
```

# Step 3: Forecast and specify predictions intervals
This is done using the `level` parameter. It takes values between 0 and 100.

```python
forecast_df = nixtla_client.forecast(
    df=df,
    h=12,
    time_col='timestamp',
    target_col="value",
    level=[80]
)
```

## Step 4: Plot predictions and intervals

```python
nixtla_client.plot(
    df=df, 
    forecasts_df=forecast_df, 
    time_col='timestamp', 
    target_col='value',
    level=[80]
)
```
